```python
import numpy as np
import pandas as pd

brfss_clean=pd.read_csv("/content/brfss_clean.csv")
uganda_clean=pd.read_csv("/content/uganda_clean.csv")
geda_clean=pd.read_csv("/content/geda_clean.csv")

print("BRFSS:", brfss_clean.shape)
print("Uganda:", uganda_clean.shape)
print("GEDA:", geda_clean.shape)
```

```python
#creating harmonized age groups

#BRFSS age groups
brfss_h=brfss_clean.copy()
brfss_h["age_group"]=np.where(
    brfss_h["age"] <40, 1,
    np.where(brfss_h["age"] <60, 2, 3)
)

#Uganda age groups
uganda_h =uganda_clean.copy()
uganda_h["age_group"]=np.where(
    uganda_h["age"] <40, 1,
    np.where(uganda_h["age"] <60, 2, 3)
)

#GEDA age groups
geda_h=geda_clean.copy()
geda_h["age_group"]=np.where(
    geda_h["age"].isin([1, 2, 3, 4]),1,
    np.where(geda_h["age"].isin([5,6]), 2,
             np.where(geda_h["age"].isin([7, 8, 9]), 3, np.nan))
)
```

```python
#creating harmonized BMI groups
#BRFSS BMI groups

brfss_h["bmi_group"]=np.where(
    brfss_h["bmi"] <25, 1,
    np.where(brfss_h["bmi"] <30, 2,
             np.where(brfss_h["bmi"] >=30, 3, np.nan))
)

#Uganda BMI groups
uganda_h["bmi_group"]=np.where(
    uganda_h["bmi"] <25, 1,
    np.where(uganda_h["bmi"] <30, 2,
             np.where(uganda_h["bmi"] >=30, 3, np.nan))
)

#GEDA BMI groups
geda_h["bmi_group"]=np.where(
    geda_h["bmi"].isin([1, 2, 3]),1,
    np.where(geda_h["bmi"].isin([4,5]), 2,
             np.where(geda_h["bmi"]== 6, 3, np.nan))
)
```

```python
final_cols =[
    "diabetes_binary",
    "age_group",
    "sex_male",
    "bmi_group",
    "hypertension",
    "current_smoker",
    "physically_active"
]

brfss_final =brfss_h[final_cols].copy()
uganda_final=uganda_h[final_cols].copy()
geda_final=geda_h[final_cols].copy()

print("BRFSS final:", brfss_final.shape)
print("Uganda final:", uganda_final.shape)
print("GEDA final:", geda_final.shape)
```

```python
#checking if the grouped variables were created correctly

for name, df in [("BRFSS", brfss_final), ("Uganda", uganda_final), ("GEDA", geda_final)]:
    print(f"\n==== {name} ====")
    print("Outcome distribution:")
    print(df["diabetes_binary"]. value_counts(dropna=False))
    print("\nMissingness:")
    print(df.isna().mean().sort_values())
    print("\nAge groups:")
    print(df["age_group"]. value_counts(dropna=False))
    print("\nBMI groups:")
    print(df["bmi_group"]. value_counts(dropna=False))
```

```python
brfss_final.to_csv("/content/brfss_final_harmonized.csv", index=False)
uganda_final.to_csv("/content/uganda_final_harmonized.csv", index=False)
geda_final.to_csv("/content/geda_final_harmonized.csv", index=False)

print("Saved all harmonization datasets.")
```
