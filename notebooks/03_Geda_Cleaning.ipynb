I have identified the main harmonization candidates for the German analysis: a diabetes diagnosis variable, a hypertension diagnosis variable, a current smoking variable, a six category BMI variable, a binary sport-participation variable, grouped sex and age.

```python
#GEDA Cleaning
geda_sub =geda[[
    "KHdiabB12",
    "KHhyp12",
    "RCstatE_k3",
    "PAbmiB_k2",
    "KAehispaq6",
    "age10B",
    "sex"
]].copy()

print(geda_sub.shape)
geda_sub.head()
```

```python
#Recoding the variables

#Diabetes: 1=yes, 2=no
geda_sub["diabetes_binary"] =np.where(
    geda_sub["KHdiabB12"]==1.0, 1,
    np.where(geda_sub["KHdiabB12"] == 2.0, 0, np.nan)

)

#Hypertension: 1=yes, 2=no
geda_sub["hypertension"] =np.where(
    geda_sub["KHhyp12"]==1.0, 1,
    np.where(geda_sub["KHhyp12"] == 2.0, 0, np.nan)

)

#Current smoker: 1=yes, 2=no
geda_sub["current_smoker"] =np.where(
    geda_sub["RCstatE_k3"]==1.0, 1,
    np.where(geda_sub["RCstatE_k3"] == 2.0, 0, np.nan)

)

#BMI
geda_sub["bmi"] =geda_sub["PAbmiB_k2"]

#Physical activity, 0 days/week -->inactive and 1-7 days/week -->active
geda_sub["physically_active"] =np.where(
    geda_sub["KAehispaq6"]== 0.0, 0,
    np.where(geda_sub["KAehispaq6"]. between (1.0, 7.0), 1, np.nan)
)

#Age: grouped age
geda_sub["age"] =geda_sub["age10B"]

#Sex: assume 1=male, 2=female
geda_sub["sex_male"] =np.where(
    geda_sub["sex"]==1.0, 1,
     np.where(geda_sub["sex"] == 2.0, 0, np.nan)

)
```

```python
geda_clean= geda_sub[[
    "diabetes_binary",
    "age",
    "sex_male",
    "bmi",
    "hypertension",
    "current_smoker",
    "physically_active"
]].copy()

print(geda_clean.shape)
print("\nOutcome distribution:")
print(geda_clean["diabetes_binary"].value_counts(dropna=False))

print("\nMissingness per variable:")
print(geda_clean.isna().mean().sort_values())

geda_clean.head()
```
